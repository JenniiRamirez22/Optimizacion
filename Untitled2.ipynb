```python
# -*- coding: utf-8 -*-
"""ejercicios-funcion-matrices.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/13S15uqK424h9wMjqDn84yZc2LkfrEZ2n
"""

import pandas as pd
import numpy as np

# =======================================================
# DEFINICIONES DE FUNCIONES
# =======================================================

#Funcion objetivo
def f(x):
    # Convertimos x en vector columna para poder multiplicar
    x_col = x.reshape(-1, 1)

    # Operación matricial
    term1 = 0.5 * (x_col.T @ Q @ x_col)
    term2 = c.T @ x_col

    return (term1 + term2).item()

#Gradiente grad = Q * x + c
def grad_f(x):
    # Convertimos x en vector columna
    x_col = x.reshape(-1, 1)

    # Calculamos el gradiente (el resultado será un vector columna 3x1)
    gradiente_columna = (Q @ x_col) + c

    # IMPORTANTE: .flatten() vuelve a aplanar el vector columna (de 3x1 a plano)
    # Se hace porque los algoritmos de optimización operan con vectores planos
    return gradiente_columna.flatten()


# Hessiana H = Q
def hessian_f(x):
    return Q

# Inicializacion de matrices
Q = np.array([[8,  3e4,  0.0],
              [3e4,  4e8,  1e10],
              [0.0,  1e10,  6e12]])

c = np.array([[11e4],
              [8e4],
              [7e6]])

def newton(x0, tol=1e-6, max_iter=100):
  historial=[]
  x = np.array(x0, dtype=float)

  k = 0

  for k in range(max_iter):
    # --- PASO 2: Calcular dirección de descenso ---
    grad=grad_f(x)
    hessian=hessian_f(x)
    try:
      # Resolvemos la direccion de descenso H * d = -g
      dir = np.linalg.solve(hessian, -grad)
    except np.linalg.LinAlgError:
      print("Error: La matriz Hessiana es singular en esta zona. No se puede invertir.")
      break

    # --- PASO 3: Establecer tamaño de paso ---
    alpha = 1.0

    # --- PASO 4: Actualizar el siguiente punto ---
    x_next = x + alpha * dir

    # --- PASO 5: Calcular norma y evaluar la condición de parada ---
    grad_next = grad_f(x_next)
    norm_grad = np.linalg.norm(grad_next)

      # Decidimos el estado del algoritmo para la tabla
    if norm_grad < tol:
        condition = "🛑 DETENIDO (Óptimo encontrado)"
    elif k + 1 >= max_iter:
        condition = "⚠️ DETENIDO (Máx. Iteraciones alcanzado)"
    else:
        condition = "🔄 SIGUE"

    # GUARDAR DATOS EN EL HISTORIAL (Formateados para que la tabla se vea limpia)
    historial.append({
            "Iter (k)": k + 1,
            "Punto Inicial": np.round(x, 5).tolist(),
            "d": np.round(dir, 5).tolist(),
            "Valor de x": np.round(x_next, 5).tolist(),
            "||∇f||": f"{norm_grad}",
            "Condición": condition
    })

    # Si la condición dictó detenerse, rompemos el bucle explícitamente aquí
    if condition != "🔄 SIGUE":
      x = x_next # Guardamos el resultado final
      break

    # --- PASO 6: Incrementar contador e iterar con el nuevo punto ---
    x = x_next

  return np.array(x), pd.DataFrame(historial)


def quasi_newton(x0, tol=1e-6, max_iter=100):
    historial = []
    x = np.array(x0, dtype=float)
    n = len(x)

    # B aproxima al HESSIANO (no al inverso). Por ende, d se resuelve o se usa B
    # Nota: Como tu fórmula usa Bk = B + ..., estás aproximando la matriz B (Hessiano).
    # Para la dirección d, lo correcto si aproximas el Hessiano es resolver B @ d = -g.
    B = np.eye(n)

    # Calculamos el gradiente inicial de partida
    g = grad_f(x)

    for k in range(max_iter):
        fx = f(x)
        g_norm = np.linalg.norm(g)

        # Evaluar criterio de parada con el gradiente actual
        if g_norm < tol:
            condition = "🛑 DETENIDO (Óptimo encontrado)"
        elif k + 1 >= max_iter:
            condition = "⚠️ DETENIDO (Máx. Iteraciones alcanzado)"
        else:
            condition = "🔄 SIGUE"

        # Paso 2: Determinar la dirección de descenso (d_k) resolviendo B @ d = -g
        # Usamos linalg.solve porque B es la aproximación del Hessiano directo
        d = np.linalg.solve(B, -g)

        # Paso 3: Tamaño de paso
        alfa = 1.0

        # Paso 4: Calcular el siguiente punto
        x_next = x + alfa * d
        g_next = grad_f(x_next) # Ahora sí existe de forma segura

        # Guardamos en el historial los datos reales de esta iteración k
        historial.append({
          "Iteración (k)": k + 1,
          "Punto Inicial": np.round(x, 6).tolist(),
          "d": np.round(d, 6).tolist(), # Corregido: 'd' en vez de 'dir'
          "x_next": np.round(x_next, 6).tolist(),
          "||∇f||": f"{g_norm:.6e}", # Corregido: Formato científico para el gradiente
          "Condición": condition
        })

        # Si se cumple la parada, rompemos el ciclo ANTES de actualizar la matriz
        if condition != "🔄 SIGUE":
            x = x_next
            break

        # --- PASO EXTRA: ACTUALIZACIÓN DE LA MATRIZ B (SR1) ---
        sk = (x_next - x).reshape(-1, 1)
        yk = (g_next - g).reshape(-1, 1)

        zk = yk - B @ sk
        denominator = zk.T @ sk

        # Salvaguarda numérica para evitar divisiones por cero
        if np.abs(denominator) > 1e-8 * np.linalg.norm(sk) * np.linalg.norm(zk):
            B = B + (zk @ zk.T) / denominator # Corregido: Guardamos en B

        # Desplazamos las variables para la siguiente iteración k+1
        x = x_next
        g = g_next

    return np.array(x), pd.DataFrame(historial)
#===========================================================
# EJECUCIÓN DE ALGORITMOS
#===========================================================

# Definimos el punto de partida inicial
x0 = np.array([0.1, 0.1, 0.1])

# Ejecutamos Newton Puro
x_opt_newton, df_newton = newton(x0)

# Ejecutamos Cuasi-Newton SR1
x_opt_quasi, df_quasi = quasi_newton(x0)


# --- REPORTE DE RESULTADOS ---
print("-" * 50)
print("             RESULTADOS")
print("-" * 50)

# Mostramos los resultados para Newton

display(df_newton)
print("-" * 50)
print(" MÉTODO DE NEWTON:")
print(f"  El minimizador x* es           : [{x_opt_newton[0]:.6f}, {x_opt_newton[1]:.6f}, {x_opt_newton[2]:.6f}]")
print(f"  La función evaluada en x*      : {f(x_opt_newton):.6e}")
print(f"  Iteraciones requeridas         : {len(df_newton) - 1}")
print("-" * 50)

# Mostramos los resultados para Cuasi-Newton SR1
display(df_quasi)
print("-" * 50)
print(" MÉTODO CUASI-NEWTON SR1:")
print(f"  El minimizador x* es           : [{x_opt_quasi[0]:.6f}, {x_opt_quasi[1]:.6f}, {x_opt_quasi[2]:.6f}]")
print(f"  La función evaluada en x*      : {f(x_opt_quasi):.6e}")
print(f"  Iteraciones requeridas         : {len(df_quasi) - 1}")
print("-" * 50)
```
